## Investigation Overview

## Summary
> Most of the loans are intermediate term loans with a repayment period of 1 to 5 years which are mostly used for debt consolidation (amount received from the new loan is used to pay off other debts). When consolidating debt, your credit is checked, which can lower or improve your credit score. Hence the normal distribution of the prosper rating. One of the biggest disadvantages of debt consolidation is that it is not accessible to everyone. From the analysis, most of the loans are given to employed persons with an income range of $25,000 and above. It is also observed that there is a steady increase in loan disbursements from 2009 till date after the global financial crisis. I am most interested in figuring out what features are best for predicting the borrower's Annual Percentage Rate and Loan Amount.

## Key Insights
> - Relationship between the main features(BorrowerAPR and Loan Amount):The BorrowerAPR helps a borrower determine the true cost of the loan. From analysis there was a negative relationship which means that one variable tends to cause a decrease in another variable. The more money you borrow, the lower your BorrowerAPR
> - Prosper Rating Effect on Relationship between BorrowerAPR and Loan Amount: ProsperRating is a significant measure of loan amount and BorrowerAPR. ProsperRating tells us how risky the borrower is. Having a good ProsperRating can help to increase a borrowerâ€™s loan amount and improve BorrowerAPR. Since most of the loans is for debt consolidation, the ProsperRating is normally distributed.
> - Term Effect on Loan Amount: There is positive correlation between Term and Loan Amount. The higher the loan, the longer the term. 
> - IncomeRange effect on Loan Amount and BorrowerAPR: The income of borrowers played a significant role in determining the loan amount which also affects the BorrowerAPR. Borrowers with good sources of income can borrow large loans.




## Dataset Overview
> The prosper loan dataset is a table of 113937 entries and 81 columns. after data wrangling and understanding the features, the cleaned dataset contains 113937 entries with 14 columns (including LoanOriginalAmount, BorrowerAPR, IncomeRange, EmploymentStatus and many others)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from datetime import datetime

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan = pd.read_csv("prosper_loan_master.csv")
```

## Distribution of Borrower APR
> The distribution of APR looks multimodal. A small peak centered at 0.1, a large peak centered at 0.2. There is also a small peak centered 0.3. Additionally, there is a very shape peak between 0.35 and 0.36. Only very few loans have APR greater than 0.43.



```python
bins = np.arange(0, df_loan.BorrowerAPR.max()+0.05, 0.01)
plt.figure(figsize=[12, 5])
plt.hist(data = df_loan, x = 'BorrowerAPR', bins = bins);
plt.xlabel('Borrower APR');
```

## Distribution of Loan Original Amount

> The very large spikes in frequency are at 4k, 5k, 8k, 10k, 15k, 20k, 25k. It means that most of the loans are multiples of 5k.

```python
# Define the figure size
plt.figure(figsize = [12, 5])


# Get the ticks for bins between [0 - maximum weight]
bins = 10 ** np.arange(3, 4.5+0.1, 0.1)

# Plot the histogram
plt.hist(data=df_loan, x='LoanOriginalAmount', bins=bins);

# The argument in the xscale() represents the axis scale type to apply.
# The possible values are: {"linear", "log", "symlog", "logit", ...}
plt.xscale('log')

# Apply x-ticks
plt.xticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4,4e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k','40k'])
plt.xlabel('Original Loan Amount (Log Transformed)');
```

## Distribution of ProsperRating

> There is a normal distribution in the ratings. The lowest risked loan is AA and highest risked loan is HR, and NC is No Credit rating.


```python
#Convert ProsperRating to category

# initiate the categorical class type with the pd.api method
ratings = ["AA", "A", "B", "C", "D", "E", "HR", "NC"]
rating_category = pd.api.types.CategoricalDtype(ratings, ordered= True)

#Convert datatype of prosperrating into the initiated category above

df_loan["ProsperRating"] = df_loan["ProsperRating"].astype(rating_category)
```

```python
plt.figure(figsize=(8,5))
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x = 'ProsperRating',color = base_color);
plt.xlabel('ProsperRating', fontsize=16)
plt.show()
```

## Distribution of IncomeRange

>  Most of the loans are given to income earners. 

```python
#Rename the values using the str.replace method
df_loan["IncomeRange"] = df_loan["IncomeRange"].str.replace("Not employed", "$0")

# initiate the categorical class type with the pd.api method
range = ["$100,000+", "$75,000-99,999", "$50,000-74,999", "$25,000-49,999", "$1-24,999", "$0", "Not displayed"]
range_category = pd.api.types.CategoricalDtype(range, ordered= True)

#Convert datatype of IncomeRange into the initiated category above

df_loan["IncomeRange"] = df_loan["IncomeRange"].astype(range_category)
```

```python
plt.figure(figsize=(12,5))
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x = 'IncomeRange',color = base_color);
plt.xlabel('IncomeRange', fontsize=16)
plt.show()
```

## Distribution of Loan Purposes
> Most of the loans are refinancing debts.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(12,5))
sb.countplot(data = df_loan, y= "ListingCategory" ,color = base_color);
plt.xlabel('Loan Purposes', fontsize=16)
plt.show()
```

## Distribution of Loans across time

> There is a significant growth in loan disbursements in recent years

```python
plt.figure(figsize=(12,5))
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x = 'ListingYear',color = base_color);
plt.xlabel('ListingYear', fontsize=16)
plt.show()
```

## Distribution of Loan Term

> Most of the loans are intermediate loans

```python
plt.figure(figsize=(8,5))
base_color = sb.color_palette()[0]
sb.countplot(data = df_loan, x = 'Term',color = base_color);
plt.xlabel('Term', fontsize=16)
plt.show()
```

## Borrower APR vs. Loan Amount
> A negative correlation means the increase in one variable tends to cause a decrease in another variable. This means that the more the loan amount, the lower the Borrower APR.

```python
# bivariate plot of LoanOriginalAmount vs. BorrowerAPR
# Scatter plot

plt.figure(figsize = [18, 6])

#PLOT ON LEFT
plt.subplot(1, 2, 1)
plt.scatter(data = df_loan, x = 'BorrowerAPR', y = 'LoanOriginalAmount', alpha = 1/10);
plt.xlabel('BorrowerAPR')
plt.yscale('log')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k'])
plt.ylabel('Original Loan Amount (Log Transformed)')


# PLOT ON RIGHT
plt.subplot(1, 2, 2)
bins_x = np.arange(0, df_loan.BorrowerAPR.max()+0.05, 0.01)
bins_y = 10 ** np.arange(3, 4.5+0.1, 0.1)
plt.hist2d(data = df_loan, x = 'BorrowerAPR', y = 'LoanOriginalAmount', cmin=0.5, cmap='viridis_r',  bins = [bins_x, bins_y])
plt.colorbar()
plt.yscale('log')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k'])
plt.xlabel('BorrowerAPR')
plt.ylabel('Original Loan Amount (Log Transformed)');

corr = np.corrcoef(df_loan['BorrowerAPR'], df_loan['LoanOriginalAmount'])
print("the correlation between BorrowerAPR and Original Loan Amount is: {}.".format(corr[0,1]))
```

## Distribution of IncomeRange across years

> Income range plays significant role in loan disbursements. From the plot, we notice that borrowers with no income ranges were not given loans after 2007 and borrowers that did not have source of income were not given loans in 2014

```python
plt.figure(figsize=(8,5))
sb.countplot(x='ListingYear', hue='IncomeRange', data=df_loan)
plt.title("Plot IncomeRange with respect to IncomeRange in the data using countplot")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5));
plt.show()
```

```python
# create a sample of 1000 
np.random.seed(2020)
samples = np.random.choice(a = df_loan.shape[0],size = 1000, replace = False)
loan_sample = df_loan.iloc[samples]
```

## Prosper Rating Effect on Relationship between BorrowerAPR and Loan Amount
> ProsperRating is a major factor in determining the loan amount and BorrowerAPR. The better your ratings, the better your APR and loan amount

```python
g = sb.FacetGrid(data = loan_sample, hue = 'ProsperRating', size = 7)
g.map(plt.scatter,'BorrowerAPR', 'LoanOriginalAmount')
g.set(yscale = 'log')
plt.ylabel('LoanOriginalAmount (Log Transformed)')
plt.xlabel('BorrowerAPR')
plt.legend(title= 'ProsperRating' , loc = "center left", bbox_to_anchor = (1, 0.5))
plt.title('BorrowerAPR and LoanOrginal Amount for ProsperRating')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4,4e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k','40k']);
```

## Term Effect on Relationship between BorrowerAPR and Loan Amount
> High value loans are attached to longer terms.

```python
g = sb.FacetGrid(data = loan_sample, hue = 'Term', size = 7)
g.map(plt.scatter,'BorrowerAPR', 'LoanOriginalAmount')
g.set(yscale = 'log')
plt.ylabel('LoanOriginalAmount (Log Transformed)')
plt.xlabel('BorrowerAPR')
plt.legend(title= 'Term' , loc = "center left", bbox_to_anchor = (1, 0.5))
plt.title('BorrowerAPR and LoanOrginal Amount for Term')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4,4e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k','40k']);
```

## Income Effect on Relationship between BorrowerAPR and Loan Amount
> The income of borrowers has a major effect on borrowerAPR and Loan Amount. The more the borrowers earn, the more loan he gets.

```python
g = sb.FacetGrid(data = loan_sample, hue = 'IncomeRange', size = 7)
g.map(plt.scatter,'BorrowerAPR', 'LoanOriginalAmount')
g.set(yscale = 'log')
plt.ylabel('LoanOriginalAmount (Log Transformed)')
plt.xlabel('BorrowerAPR')
plt.legend(title= 'IncomeRange' , loc = "center left", bbox_to_anchor = (1, 0.5))
plt.title('BorrowerAPR and LoanOrginal Amount for ProsperRating')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4,4e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k','40k']);
```

```python
# convert "LoanStatus" to category type
df_loan[[ "LoanStatus"]] = df_loan[["LoanStatus"]].astype("category")
```

## Relationship between BorrowerAPR and ProsperRating and loan status
> There is an increase in BorrowerAPR as the loan gets more risky. The Defaulted loans in the least riskest loan has a higher APR.

```python
plt.figure(figsize = [12,6])

ax = sb.pointplot(data = df_loan, x = 'ProsperRating', y = 'BorrowerAPR', hue = 'LoanStatus', dodge = 0.2, linestyles="", 
                  palette="prism")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5))
plt.xticks(rotation= 0)
plt.show();
```

## Relationship between ProsperRating, Loan Amount and Loan Status

```python
plt.figure(figsize = [12,6])

ax = sb.pointplot(data = df_loan, x = 'ProsperRating', y = 'LoanOriginalAmount', hue = 'LoanStatus', dodge = 0.2, linestyles="", 
                  palette="prism")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5))
plt.yscale('log')
plt.yticks([1e3, 2e3,3e3,4e3, 5e3, 8e3,1e4, 15e3,2e4,3e4], ['1k', '2k','3k','4k' ,'5k','8k', '10k','15k','20k','30k']);
plt.xticks(rotation= 0)
plt.show();
```
